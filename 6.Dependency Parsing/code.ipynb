Using spaCy

```python
import spacy
import nltk
# Load English tokenizer, POS tagger, parser, NER, and word vectors
nlp = spacy.load("en_core_web_sm")

#sentence 
text = "The quick brown fox jumps over the lazy dog."

# process the sentence
doc=nlp(text)

# Print dependencies
print(f"{'Token':10} {'Dep':15} {'Head':10} {'POS':10}")
print("-" * 50)
for token in doc:
    print(f"{token.text:10} {token.dep_:15} {token.head.text:10} {token.pos_}")
```

### (Optional)  view the dependency tree

```python
for token in doc:
    if token.dep_ == "ROOT":
        print(f"Root verb: {token.text}")

for token in doc:
    if token.dep_ == "nsubj":
        print(f"Subject: {token.text} (head: {token.head.text})")

for token in doc:
    if token.dep_ == "prep":
        print(f"Preposition: {token.text} (attached to: {token.head.text})")

for token in doc:
    if token.dep_ == "pobj":
        print(f"Object of preposition: {token.text} (from: {token.head.text})")

for token in doc:
    if token.dep_ == "amod":
        print(f"Adjective: {token.text} (modifies: {token.head.text})")
```
