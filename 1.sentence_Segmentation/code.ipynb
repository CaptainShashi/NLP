Option 2: Regex-Based Segmentation

```python
import re   # regular expression  used for searching, splitting, or matching patterns in text
text="Hello! My name is Shashikumar. I love NLP. Do you?"
output=re.split(r'(?<=[.!?])\s+', text)                    # re.split(pattern, string)  eg.. (?<=\$) means: "Look behind for a $ symbol"
print(output)

# Might break sentences at Mr., Dr., U.S.   Might miss actual sentence boundaries
```

Using Spcay 

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Dr. Smith went to the U.S. He works in A.I. research."
doc = nlp(text)

sentences = [sent.text for sent in doc.sents]
print(sentences)
```
