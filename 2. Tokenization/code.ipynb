using split()

```python
text='hey! my name is shashikumar, i work at muffins AI'
words=text.split()
print(words)
```

using re - regular expression

```python
import re
text='hey! my name is shashikumar, i work at muffins AI'

words=re.findall(r"\w+|[^\w\s]", text)
print(words)
```

Punkt tokenizer

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

text = "Hello there! How're you doing?"
words = word_tokenize(text)
print(words)
# Output: ['Hello', 'there', '!', 'How', "'re", 'you', 'doing', '?']
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Hello there! How're you doing?")

words = [token.text for token in doc]
print(words)
# Output: ['Hello', 'there', '!', 'How', "'re", 'you', 'doing', '?']
```
